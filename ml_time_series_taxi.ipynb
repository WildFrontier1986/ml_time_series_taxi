### Проверка стационарности временного ряда

<b>Интегрированный временной ряд</b> — нестационарный временной ряд, разности некоторого порядка от которого являются стационарным временным рядом.

Наличие единичных корней в авторегрегрессионной модели временного ряда эквивалентно понятию интегрированности временного ряда.

Таким образом, признаком нестационарного временного ряда является наличие единичного корня. 

<b>Нулевая гипотеза</b> расширенного теста <code>Дики-Фуллера</code> состоит в том, что существует единичный корень; <b>альтернативная гипотеза</b> - единичного корня нет. Если p-value выше критического размера, мы не можем отрицать наличие единичного корня.

```python
# Запуск расширенного теста Дики-Фуллера
results_adf = adfuller(x=resampled_1h, regression='c', autolag='AIC')
```

```python
# Вывод результатов расширенного теста
# Дики-Фуллера на стационарность
print(
    f'Тестовая статистика ADF: {results_adf[0]:.3f}\n',
    f'Уровень статистической значимости p-value: {results_adf[1]:.2%}\n',
    'Критические значения тестовой статистики при уровне статистической значимости:\n',
    f'1%: {results_adf[4]["1%"]:.3f}\n',
    f'5%: {results_adf[4]["5%"]:.3f}\n',
    f'10%: {results_adf[4]["10%"]:.3f}\n',
    sep=''
)
if (results_adf[0] > results_adf[4]["5%"] or
   results_adf[1] > .05):
    print('Нулевая гипотеза не отвергнута – временной ряд не стационарен')
else:
    print('Нулевая гипотеза отвергнута – временной ряд стационарен')
```

### Создание функции, формирующей датафреймы с признаками

```python
# Создание функции, формирующей датафреймы с признаками
# 1. Объявление функции frankenstein()
def frankenstein(
    heart, 
    base_column, 
    n_shift, 
    win_size,
    test_size_income=.1,
    not_stationary=False
):
    # 2. Логичесое условие для обработки ряда
    if not_stationary == True:
        # 2.1. Создание колонки стационарного временного ряда
        heart['stationary'] = heart[base_column] - heart[base_column].shift()
        # 2.2. Вывод стационарного временного ряда
        plt.figure(
            figsize=(20, 6),
            facecolor='steelblue'
        )
        sns.lineplot(
            data=heart['stationary'],
            palette=sns.color_palette(palette='hls', n_colors=1),
            legend=False,
            ci=None
        )
        plt.title('стационарный временной ряд'.upper())
        plt.grid(visible=True)
        plt.show()
        # 2.3. Формирование колонок с календарными признаками
        heart['month'] = heart.index.month
        heart['day_of_week'] = heart.index.dayofweek
        heart['hour'] = heart.index.hour
        # 2.4. Формирование колонок с отстающими признаками
        for n in range(1, n_shift + 1):
            heart[f'shift_{n}'] = heart['stationary'].shift(n)
        # 2.5. Формирование колонки со скользящим средним
        heart['rolling_mean'] = heart['stationary'].shift().rolling(win_size).mean()
        # 2.6. Формирование колонки со скользящим стандартным отклонением
        heart['rolling_std'] = heart['stationary'].shift().rolling(win_size).std()
    # 3. Логичесое условие для обработки ряда
    else:
        # 3.1. Формирование колонок с календарными признаками
        heart['month'] = heart.index.month
        heart['day_of_week'] = heart.index.dayofweek
        heart['hour'] = heart.index.hour
        # 3.2. Формирование колонок с отстающими признаками
        for n in range(1, n_shift + 1):
            heart[f'shift_{n}'] = heart[base_column].shift(n)
        # 3.3. Формирование колонки со скользящим средним
        heart['rolling_mean'] = heart[base_column].shift().rolling(win_size).mean()
        # 3.4. Формирование колонки со скользящим стандартным отклонением
        heart['rolling_std'] = heart[base_column].shift().rolling(win_size).std()
    # 4. Удаление объектов с пропусками в признаках
    heart.dropna(inplace=True)
    # 5. Сброс индексов
    heart.reset_index(drop=True, inplace=True)
    # 6. Формирование исходных выборок с признаками
    # (вопросами) и целевыми признаками (ответами)
    if not_stationary == True:
        x = heart.drop(columns=[base_column, 'stationary'], axis=1)
        y = heart['stationary']
    else:
        x = heart.drop(columns=[base_column], axis=1)
        y = heart[base_column]
    # 7. Формирование обучающих и тестовых выборок
    x_train, x_test, y_train, y_test = train_test_split(
        x,
        y,
        test_size=test_size_income,
        shuffle=False,
        random_state=777
    )
    # 8. Возвращение функции
    return heart, x_train, x_test, y_train, y_test
```

### Создание функции, вычисляющей значение RMSE

```python
# Создание функции, возвращающей целевые признаки
# от разницы рядов к значению ряда и вычисляющей
# по полученным значениям метрику RMSE
# 1. Объявление функции back_future_rmse()
def back_future_rmse(
    y_true,
    y_pred,
    not_stationary=False,
    data_income=df_full['num_orders']
):
    # 2. Логичесое условие для расчёта
    # RMSE нестационарного ряда
    if not_stationary == True:
        # 2.1. Формирование Series с предсказаниями
        # из массива np.array
        y_pred = pd.Series(y_pred)
        # 2.2. Перезапись индексов в Series с предсказаниями
        y_pred.index = y_true.index
        # 2.3. Возвращение от разницы рядов к значению ряда
        y_pred_back = (
            data_income[y_true.index.min() - 1:y_true.index.max() + 1].shift() +
            y_pred
        )
        y_true_back = (
            data_income[y_true.index.min() - 1:y_true.index.max() + 1].shift() +
            y_true
        )
        # 2.4. Удаление объектов с пропусками
        y_pred_back.dropna(inplace=True)
        y_true_back.dropna(inplace=True)    
        # 2.5. Вычисление квадратного корня из
        # средней квадратичной ошибки
        rmse = mean_squared_error(
            y_true_back,
            y_pred_back,
            squared=False
        )
    # 3. Логичесое условие для расчёта
    # RMSE стационарного ряда
    else:
        rmse = mean_squared_error(
            y_true,
            y_pred,
            squared=False
        )
    # 7. Возвращение функции
    return rmse
```

### Создание метрики качества RMSE

```python
# Создание метрики качества RMSE для аргумента
# 'scoring' в кросс-валидации
rmse_scorer = make_scorer(
    back_future_rmse,
    greater_is_better=False
)
```

### Кросс-валидация признаков временных рядов

<b>Для обучения и тестирования моделей буду использовать поиск по сетке параметров с кросс-валидацией.<br>
    Так как предсказания выполняются для временных рядов, то кросс-валидация должна выполняться последовательно без перемешивания признаков, а признаки из валидационной/тестовой выборки не должны попадать в обучающую выборку.<br>
    Для формирования массивов последовательных индексов обучающей и валидационной/тестовой выборок применю <code>TimeSeriesSplit()</code>. Количество подвыборок задам таким образом, чтобы размер валидационной был примерно равен размеру глобальной тестовой выборки. Кросс-валидатор временного ряда передам в параметр <code>cv</code> функции <code>GridSearchCV</code>.</b>
